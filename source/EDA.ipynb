```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/processed/fraudTrain.csv')
df.head(5)
```

```python
fraud_counts = df["is_fraud"].value_counts().sort_index()
print(fraud_counts)
plt.bar(['Not Fraud', 'Fraud'], fraud_counts)
plt.title('Number of Fraudulent vs Non-Fraudulent Transactions in fraudTrain')
plt.ylabel('Count')
plt.xlabel('Transaction Type')
plt.show()
```
